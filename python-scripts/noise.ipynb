```python
from matplotlib import pyplot as plt 
import numpy as np
import pandas as pd
```

```python
acc = pd.read_csv('../config/log/Graph2.txt')
f, ax = plt.subplots(1, 2, figsize=(20, 5))

acc.plot(x='time', y=' Quad.IMU.AX', ax=ax[0])
acc[' Quad.IMU.AX'].plot.kde(ax=ax[1])

print(f"Mean ACC.x: {acc[' Quad.IMU.AX'].mean():.2f}")
print(f"Std ACC.x: {acc[' Quad.IMU.AX'].std():.2f}")
```
